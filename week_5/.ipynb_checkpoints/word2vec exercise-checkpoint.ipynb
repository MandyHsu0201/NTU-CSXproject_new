## word2vec 用途:
選取訓練後的單詞向量的其中任意三個維度，放到座標圖展示，會發現語意相似的詞彙在空間中的位置相近(位移向量也相似)，語意無關的慈則較遠。可用來對單詞賀句子進行更加泛化的分析

## word2vec 應用 1:給廣告主推薦用戶，
如網頁a1,a2,a3,a4的距離分別是0.3,0.4,0.5,0.6，則距離無意義，但a2與a1的相似度比a3,a4要大(使用cos function來衡量)。所以可以推論成瀏覽a1的用戶會瀏覽a2,a3,a4的機率很高，同樣瀏覽a2,a3,a4的用戶也可能會瀏覽a1。
如a1是匹克体育用品公司在媒体公司A上的官网，a2是湖人队比赛数据页，a3是热火队的灌水讨论区，a4是小牛队的球员讨论区，則匹克体育用品的廣告主想要找經常瀏覽公司官網的用戶，也就相當於找瀏覽a2,a3,a4的用戶

## word2vec 應用 2:對ctr(Click through rate點擊率,廣告點選次數除以Impress(廣告曝光次數))預估模型的幫助
參考互联网广告综述之点击率特征工程https://blog.csdn.net/mytestmy/article/details/19088827
廣告特徵:性別,年齡,手機平台,地域,廣告反饋ctr(當前的廣告正在投放，已經投放了一部分即可當作其點擊率),廣告圖片大小,前景色,背景色

若這個廣告是新上線的，則可以考慮使用跟他類似的廣告來估計點擊率。
若將所有網頁做分群，則可使用歷史投放數據及及時數據來預估這個流量對這群(c1如天貓掏寶亞馬遜)的點擊率。
如果投稿投放量夠，直接估計，否則就用這群的點擊率來替代

ex:广告的反馈ctr，用户年龄，性别三个特征
(1)離散化(分群的意思) 1:廣告反饋ctr，2:age1,...11:age10 (20歲,30歲看的廣告沒啥差異)
(2)交叉              本來12,13是要設男女的，但有些廣告(體育，化妝)設男女就看不出來，12改成該廣告用戶在男性用戶上面的點擊率(變成實數)

(3)連續特徵變離散特徵(所有特徵皆要執行離散化，優點:离散化的特征能拟合数据中的非线性关系，取得比原有的连续特征更好的效果，而且在线上应用时，无需做乘法运算，也加快了计算ctr的速度。) 
假設CTR服從lognormal(yahoo是假設服從Beta)，則點擊率在不同區間應該考慮不同的權重係數才是，實現的方式就是將點擊率分區間，每個區間為一個新的特徵。也就是當點擊率在0.2時(高點)，點擊率0.25的廣告就是比0.2的好，但在1.0時，1.05與1.0的廣告效果是差不多的。實現離散化的作法:將1:廣告反饋ctr做排序後分10等份，即變成10個特徵。

NOTE:比如男性資料太少時，计算广告与性别交叉的ctr是很不准确的，需要对这个特征进行修正，参考《广告点击率的贝叶斯平滑》。但離散化會造成特徵過多，則可考慮用L1正規化，可參考《从广义线性模型到逻辑回归》《OWL-QN算法》和《在线学习算法FTRL》。
